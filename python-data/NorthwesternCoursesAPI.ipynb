```python
params = {'key': 'JC3v77yRGJzdrHNn'}
response = requests.get('http://api.asg.northwestern.edu/schools/?key=YOUR_KEY', params=params)
schoolarray = json.loads(response.text)
len_schoolarray = len(schoolarray)
```

```python
#params = {'key': 'JC3v77yRGJzdrHNn', 'term':4730}
#response = requests.get('http://api.asg.northwestern.edu/subjects/?key=YOUR_KEY', params=params)


for i in range(len_schoolarray):
    schoolsymbol = schoolarray[i].get('symbol')
    params = {'key': 'JC3v77yRGJzdrHNn', 
              'term':4760, 
              'school' : schoolsymbol}
    response = requests.get('http://api.asg.northwestern.edu/subjects/?key=YOUR_KEY', params=params)
    schoolarray[i]['subjects'] = json.loads(response.text)

pprint(schoolarray)
```

```python
detailedschoolarray = schoolarray.copy()
```

```python
for schoolindex in range(len_schoolarray):
    
    school = detailedschoolarray[schoolindex]
    subjectarray = school.get('subjects')
    len_subjectarray = len(subjectarray)
    
    for subjectindex in range(len_subjectarray):
        
        params = {'key': 'JC3v77yRGJzdrHNn', 
                  'term': 4760,
                  'subject' : subjectarray[subjectindex].get('symbol')}
        response = requests.get('http://api.asg.northwestern.edu/courses/details/?key=YOUR_KEY', params=params)
        coursesarray = json.loads(response.text)
        #
        detailedschoolarray[schoolindex]['subjects'][subjectindex]['courses'] = [] 
        existingcourse = []
        for course in coursesarray:
            
            if(course['title'] not in existingcourse):
                existingcourse.append(course['title'])
                course = {'id' : course['id'],
                          'title' : course['title'],
                          'term' : course['term'],
                          'school' : course['school'], 
                          'catalog_num' : course['catalog_num'],
                          'class_num': course['class_num'],
                          'course_id': course['course_id'],
                          'subject': course['subject']}
                detailedschoolarray[schoolindex]['subjects'][subjectindex]['courses'].append(course)
            
#for full courses data, comment out evreything below the above comment symbol, and replace with
            
                
        
        
```
